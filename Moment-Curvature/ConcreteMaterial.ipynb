```python

from openseespy.opensees import *
import opsvis as opsv

import numpy as np

%matplotlib ipympl
import matplotlib.pyplot as plt

import math
import pandas as pd

# remove existing model
wipe()
plt.close('all')

plt.rcParams['figure.constrained_layout.use'] = True

fpcc = -7.89
ecpeak = -0.0073
e_cu = -0.020

# Unconfined
fpc = -5.2
fpcu = 0
eco = -0.0026
esp = -0.005

Econc = 4347
# -------------------------------------------------------------------------
# Core concrete (confined)
uniaxialMaterial("Concrete04", 1, fpcc, ecpeak, e_cu, Econc, 0.0, 0.0, 0.1)
# Cover concrete (unconfined)
#uniaxialMaterial("Concrete04", 2, fpc, eco, esp, Econc, 0.0, 0.0, 0.1)
#uniaxialMaterial("Concrete02", 2, fpc, eco, fpcu, esp, 0.1, 0.0, 0.0)
uniaxialMaterial("Concrete01", 2, fpc, eco, 0., esp)

strain_1 = np.arange(0, e_cu, -0.0001)
testUniaxialMaterial(1)
stress_1 = []
for strain_i in strain_1:
    setStrain(strain_i)
    stress_1.append(getStress())


testUniaxialMaterial(2)
stress_2 = []
for strain_i in strain_1:
    setStrain(strain_i)
    stress_2.append(getStress())

# Plot Stress-Strain relationships
# ----------------------------------------
plt.figure(figsize=(10,5))

ecu_min = e_cu*1.05
fc_min = fpcc*1.05

# Major ticks every 20, minor ticks every 5
major_xticks = np.arange(ecu_min, 0, (0 - ecu_min)/10)
minor_xticks = np.arange(ecu_min, 0, (0 - ecu_min)/50)
major_yticks = np.arange(fc_min, 0, (0 - fc_min)/10)
minor_yticks = np.arange(fc_min, 0, (0 - fc_min)/50)    

ax = plt.axes()
ax.set_xticks(major_xticks)
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
ax.set_xticks(minor_xticks, minor=True)
ax.set_yticks(major_yticks)
ax.set_yticks(minor_yticks, minor=True)

current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:.1f}'.format(x) for x in current_values])

current_values = plt.gca().get_xticks()
plt.gca().set_xticklabels(['{:.4f}'.format(x) for x in current_values])

ax.grid(which='both')
ax.grid(which='minor', alpha=0.2)
ax.grid(which='major', alpha=0.5)
#ax.fill_between(curvature, moment, alpha = 0.2)

plt.xlim(0, ecu_min)
plt.ylim(0, fc_min)

plt.xlabel('Strain [--]')
plt.ylabel('Stress [ksi]')
plt.title('Reinforcement Stress - Strain Relationship')
plt.plot(strain_1, stress_1, marker = 'o', markersize=2)
plt.plot(strain_1, stress_2, marker = 'o', markersize=2)

plt.xticks(rotation=90)
plt.show()

# ----------------------------------------
# Display Results Table:
# ----------------------------------------
ResultMatrix = {'Strain [--]': np.round(strain_1,4), 
                'Stress [ksi]': np.round(stress_1,2)}

df = pd.DataFrame(ResultMatrix, columns = ['Strain [--]', 'Stress [ksi]'])
    
pd.set_option('display.max_rows', None)     # Set to None to display all rows
pd.set_option('display.max_columns', None)  # Set to None to display all columns
   
display(df)

wipe()

```
